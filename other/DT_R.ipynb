# **二、“量”的預測模式(迴歸)：Decision Tree**

# **決策樹(Decision Tree)**

```
import pandas as pd
df=pd.read_excel('Exercise1(RA).xlsx')
```

```
import pandas as pd
import numpy as np

#準備X & y array
X = df[["education_years","working_years"]] 
y = df["salary"]  
```

```
#將資料分成訓練組及測試組
from sklearn.model_selection import train_test_split

#test_size代表測試組比例。random_state代表設定隨機種子，讓測試結果可被重複
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=333)
```

```
from sklearn.tree import DecisionTreeRegressor
Reg = DecisionTreeRegressor(criterion="mse",random_state=0)#,max_depth=5
Reg.fit(X_train, y_train)
```

```
#計算R_Square,
R2_train = Reg.score(X_train,y_train)
R2_test = Reg.score(X_test,y_test)

print('train_R2=%.3f, test_R2=%.3f'%(R2_train,R2_test))
```

```
#計算RMSE (可改用from sklearn.metrics import mean_squared_error, r2_score)
ypredict1=Reg.predict(X_train)
ypredict2=Reg.predict(X_test)
MSE1= np.mean((y_train - ypredict1)** 2)
RMSE1=MSE1**0.5
MSE2= np.mean((y_test - ypredict2)** 2)
RMSE2=MSE2**0.5
print('train_RMSE=%.3f,test_RMSE=%.3f'%(RMSE1,RMSE2))
```
