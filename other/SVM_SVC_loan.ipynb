開啟分析資料檔，並進行資料預處理

```
import pandas as pd
df=pd.read_excel('loan3(na).xlsx')
```

```
df.head(5)
#loan=2表示違約
```

```
df.isnull().any() #尋找空值變數
```

```
df1=df.dropna()
X=df1[['Industry_type','Company_type','Beginning_time','Bank_relationship','Capital_increase','Age','Education','Credit']]
y=df1['loan'] 
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

SVM 應用於分類之預測分析:SVC

第一種: LinearSVC

```
from sklearn import svm 
clf1 = svm.LinearSVC(max_iter=1000,C=1) 
clf1.fit(X_train,y_train) 
```

```
clf1_tpred= clf1.predict(X_train)
accuracy=clf1.score(X_train,y_train)
print("accuracy:",accuracy)
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
print("第一種訓練之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_train,clf1_tpred))
print("======================================================\n")
print(classification_report(y_train,clf1_tpred))
```

```
clf1_pred= clf1.predict(X_test)
print("第一種測試之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_test,clf1_pred))
print("======================================================\n")
print(classification_report(y_test,clf1_pred))
```

第二種: kernel = 'linear'

```
from sklearn import svm 
clf2 = svm.SVC(kernel = 'linear',C=1) 
clf2.fit(X_train,y_train) 
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
clf2_tpred= clf2.predict(X_train)
print("第二種訓練之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_train,clf2_tpred))
print("======================================================\n")
print(classification_report(y_train,clf2_tpred))
```

```
clf2_pred= clf2.predict(X_test)
print("第二種測試之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_test,clf2_pred))
print("======================================================\n")
print(classification_report(y_test,clf2_pred))
```

第三種: kernel = 'poly'

```
from sklearn import svm 
clf3 = svm.SVC(kernel = 'poly',degree=3, gamma="auto",C=1) 
clf3.fit(X_train,y_train) 
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
clf3_tpred= clf3.predict(X_train)
print("第三種訓練之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_train,clf3_tpred))
print("======================================================\n")
print(classification_report(y_train,clf3_tpred))
```

```
clf3_pred= clf3.predict(X_test)
print("第二種測試之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_test,clf3_pred))
print("======================================================\n")
print(classification_report(y_test,clf3_pred))
```

第四種: kernel = 'rbf'高斯轉換

```
from sklearn import svm 
clf4 = svm.SVC(kernel = 'rbf',gamma=0.8,C=1) 
clf4.fit(X_train,y_train) 
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
clf4_tpred= clf4.predict(X_train)
print("第四種訓練之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_train,clf4_tpred))
print("======================================================\n")
print(classification_report(y_train,clf4_tpred))
```

```
clf4_pred= clf4.predict(X_test)
print("第四種測試之預測力報告\n")
print("confusion_matrix: \n",confusion_matrix(y_test,clf4_pred))
print("======================================================\n")
print(classification_report(y_test,clf4_pred))
```
